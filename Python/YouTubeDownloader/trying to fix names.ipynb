```python
import os
from mutagen.easyid3 import EasyID3
from mutagen.mp3 import MP3

# Directory containing the audio files
dir_path = r"C:\Users\Usuario\Music\Musica JB\archivos de musica - Copy"

# File to store error logs
error_log_file = "error_log.txt"
error_files = []

# Function to sanitize file names
def sanitize_filename(filename):
    # Replace invalid characters with underscores
    invalid_chars = ['<', '>', ':', '"', '/', '\\', '|', '?', '*']
    for char in invalid_chars:
        filename = filename.replace(char, '_')
    return filename

# Iterate over all files in the directory
for file_name in os.listdir(dir_path):
    file_path = os.path.join(dir_path, file_name)

    # Skip directories, only process .mp3 files
    if not os.path.isfile(file_path) or not file_name.lower().endswith('.mp3'):
        continue

    try:
        # Load the MP3 file
        audio = MP3(file_path, ID3=EasyID3)

        # Extract "Title" and "Contributing Artists" metadata
        title = audio.get("title", [None])[0]
        contributing_artists = audio.get("artist", [None])[0]

        # If both attributes exist, rename the file
        if title and contributing_artists:
            # Sanitize the new file name
            sanitized_title = sanitize_filename(title)
            sanitized_contributing_artists = sanitize_filename(contributing_artists)
            new_file_name = f"{sanitized_contributing_artists} - {sanitized_title}.mp3"
            new_file_path = os.path.join(dir_path, new_file_name)

            # Avoid overwriting files with the same name
            if not os.path.exists(new_file_path):
                os.rename(file_path, new_file_path)
                print(f"Renamed: {file_name} -> {new_file_name}")
            else:
                print(f"Skipped (file already exists): {new_file_name}")
        else:
            print(f"Skipped (missing metadata): {file_name}")

    except Exception as e:
        error_message = f"Error processing file {file_name}: {e}"
        print(error_message)
        error_files.append(error_message)

# Write all errors to a log file
if error_files:
    script_dir = os.path.dirname(os.path.abspath(__file__))
    log_file_path = os.path.join(script_dir, error_log_file)
    with open(log_file_path, "w", encoding="utf-8") as log_file:
        log_file.write("\n".join(error_files))
    print(f"Errors logged in {log_file_path}")
```
